```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[10,20,30,40])
plt.xlabel('n value')
plt.ylabel("f(n) value")
plt.title("a simple plot")
plt.figure(figsize=(100,5))
plt.show()
```

```python
import numpy as np
a = np.arange(1,11)
b = a*10
x = np.arange(1,5,1)
y=x**2
plt.figure(figsize=(15,5))
plt.plot(a,b,'bo',x,y,'g^')
plt.title('simple')
```

```python
x = np.arange(1,5,1)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
a = np.arange(0,10,0.01)
b=np.sin(a)
plt.plot(a,b)
plt.show()
```

```python
months = ['jan','feb','mar','april']
values = np.arange(1,5)
a = ['summer','winter','rainy','automn']
temp = [35,15,20,28]
plt.bar(a,temp,color='blue')
plt.bar(months,values,color='green')
plt.figure()
plt.barh(months,values,color='green')
plt.show()
```
